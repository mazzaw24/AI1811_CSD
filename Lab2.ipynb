```python
#Q1
class EmptyStackException(Exception):
    def __init__(self, message="Stack is Empty"):
        self.message = message
        super().__init__(self.message)
        
class Stack:
    def __init__(self):
        self.items = []
        
    def isEmpty(self): #1
        return len(self.items) == 0

    def clear(self): #2
        self.items = []

    def push(self, item): #3
        self.items.append(item)

    def pop(self): #4
        if not self.isEmpty():
            popped_value = self.items.pop(-1)
            return popped_value
        else: 
            raise EmptyStackException()       
        
            
    def top(self):
        if not self.isEmpty():
            top = self.items[-1]
            print(f"Top: {top}")
        else:
            raise EmptyStackException()
            
    def traverse(self): #6
        if not self.isEmpty():
            a = self.items
            print(a)
        
def decimal_to_binary(decimal_number):
    binary_stack = Stack()

    while decimal_number > 0:
        remainder = decimal_number % 2
        binary_stack.push(remainder)
        decimal_number //= 2

    binary_representation = ""
    while not binary_stack.isEmpty():
        binary_representation += str(binary_stack.pop())

    reversed_binary_representation = ''.join(reversed(binary_representation))
    return reversed_binary_representation


s = Stack()
s.traverse()
t = s.isEmpty()
print(t)
print()

try:
    s.pop()  
except EmptyStackException as e:
    print(f"{e}")

print()

try:
    s.top()
except EmptyStackException as e:
    print(f"{e}")
    
print()

s.push(4)
s.push(5)
s.push(1)
s.push(3)
s.traverse()

print()

popped = s.pop()
print(f"Popped: {popped}")

s.traverse()
try:
    s.top()
except EmptyStackException as e:
    print(f"{e}")

decimal_number = 321
binary_representation = decimal_to_binary(decimal_number)
print(f"\nBinary representation of {decimal_number}:", binary_representation)
```

```python
#Q2
class EmptyQueueException(Exception):
    def __init__(self, message="Queue is Empty"):
        self.message = message
        super().__init__(self.message)
class Queue:
    def __init__(self):
        self.items = []
        
    def isEmpty(self): #1
        return len(self.items) == 0

    def clear(self): #2
        self.items = []

    def enqueue(self, item): #3
        self.items.insert(0,item)

    def dequeue(self): #4
        if not self.isEmpty():
            popped_value = self.items.pop(0)
            return popped_value
        else: 
            raise EmptyQueueException()       
        
    def first(self): #5
        if not self.isEmpty():
            top = self.items[0]
            print(f"Top: {top}")
        else:
            raise EmptyQueueException()
            
    def traverse(self): #6
        if not self.isEmpty():
            a = self.items
            print(a)
def real_to_binary(real_number):
    queue = Queue()
    binary_representation = "0."


    decimal_places_limit = 10
    for _ in range(decimal_places_limit):
        real_number *= 2
        integer_part = int(real_number)
        queue.enqueue(integer_part)
        binary_representation += str(integer_part)
        real_number -= integer_part

    return binary_representation           
q = Queue()

try:
    de = q.dequeue()  
    print(f"Popped: {de}")
except EmptyQueueException as e:
    print(f"{e}")
    
try:
    q.first()
except EmptyQueueException as e:
    print(f"{e}")
    
q.enqueue(9)
q.enqueue(1)
q.enqueue(3)
q.enqueue(2)
q.enqueue(8)
q.traverse()
de = q.dequeue()  
print(f"Popped: {de}")
q.traverse()

try:
    q.first()
except EmptyQueueException as e:
    print(f"{e}")

real_number = 0.123
if real_number >= 1:
    print("Number must be less than 1")
else:
    binary_representation = real_to_binary(real_number)
    print(f"\nBinary representation of {real_number}:", binary_representation)




```

```python
#Q5
class EmptyStackException(Exception):
    def __init__(self, message="Stack is Empty"):
        self.message = message
        super().__init__(self.message)
class EmptyQueueException(Exception):
    def __init__(self, message="Queue is Empty"):
        self.message = message
        super().__init__(self.message)   
        
class Computer:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name
    
class Stack:
    def __init__(self):
        self.items = []
        
    def isEmpty(self): #1
        return len(self.items) == 0

    def clear(self): #2
        self.items = []

    def push(self, item): #3
        self.items.append(item)

    def pop(self): #4
        if not self.isEmpty():
            popped_value = self.items.pop(-1)
            return popped_value
        else: 
            raise EmptyStackException()       
        
            
    def top(self):
        if not self.isEmpty():
            return self.items[-1]
        else:
            raise EmptyStackException()
            
    def traverse(self): #6
        if not self.isEmpty():
            a = [str(item) for item in self.items]
            print(a)
            
class Queue:
    def __init__(self):
        self.items = []
        
    def isEmpty(self): #1
        return len(self.items) == 0

    def clear(self): #2
        self.items = []

    def enqueue(self, item): #3
        self.items.insert(0,item)

    def dequeue(self): #4
        if not self.isEmpty():
            popped_value = self.items.pop(0)
            return popped_value
        else: 
            raise EmptyQueueException()       
        
    def first(self): #5
        if not self.isEmpty():
            top = self.items[0]
            print(f"Top: {top}")
        else:
            raise EmptyQueueException()
            
    def traverse(self): #6
        if not self.isEmpty():
            a = [str(item) for item in self.items]
            print(a)

def main():
    
    print("Testing Stack:")
    stack = Stack()
    stack.push(Computer("Desktop-1"))
    stack.push(Computer("Laptop-1"))
    stack.push(Computer("Server-1"))

    stack.traverse()

    popped_value = stack.pop()
    print(f"Popped value: {popped_value}")

    print("Top of the stack:", stack.top())

    stack.clear()
    print("Is stack empty now?", stack.isEmpty())

    print("\n--------------------------------\n")

   
    print("Testing Queue:")
    queue = Queue()
    queue.enqueue(Computer("Desktop-2"))
    queue.enqueue(Computer("Laptop-2"))
    queue.enqueue(Computer("Server-2"))

    queue.traverse()

    dequeued_value = queue.dequeue()
    print(f"Dequeued value: {dequeued_value}")

    print("First of the queue:", queue.first())

    queue.clear()
    print("Is queue empty now?", queue.isEmpty())

if __name__ == "__main__":
    main()
```
